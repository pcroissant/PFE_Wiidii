```
#IMPORTS

import json
import requests
import numpy as np
from tqdm import trange
import os

#NLPAUG IMPORTS
!pip install nlpaug
!pip install transformers

import nlpaug.augmenter.word as naw

import tensorflow as tf



#LOAD DATASET
with open ('/content/drive/MyDrive/IA/PFE - Wiidii/2_Workspace/code lo/dataset.json', 'r', encoding='utf-8', errors='ignore') as dataset_file:
   dataset = json.loads(dataset_file.read())


#PARAMS
NUMBER_OF_AUG = 1
dataset_aug = {}

#INITIALIZE AUGMENTERS
word_aug = naw.ContextualWordEmbsAug(model_path='camembert-base', action="substitute", aug_max=1)
#word_aug = naw.ContextualWordEmbsAug(model_path='bert-base-multilingual-cased', action="substitute", aug_max=1)
wordnet_aug = naw.SynonymAug(aug_src='wordnet', aug_max=1)


#DATA AUGMENTATION
for i in trange(len(dataset)):
    dataset_aug[i] = dataset[i]
    sentence = dataset[i].get("sentence")

    wordswap_aug = word_aug.augment(sentence, n= NUMBER_OF_AUG)
    syn_aug = wordnet_aug.augment(sentence, n= NUMBER_OF_AUG)
    print(sentence, wordswap_aug, syn_aug)


    dataset_aug[i]["wordswap_aug"] = wordswap_aug
    dataset_aug[i]["wordnet_aug"] = syn_aug


#SAVE AUGMENTED DATASET
with open('/content/drive/MyDrive/IA/PFE - Wiidii/3_Data_aug/dataset_bert_wordnet.json', "w") as dataset_aug_file:
    json.dump(dataset, dataset_aug_file, ensure_ascii=False)
```
